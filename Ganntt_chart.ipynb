```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import plotly.express as px
from datetime import datetime, timedelta

# Function to calculate working days excluding weekends
def calculate_working_days(start_date, end_date):
    current_date = start_date
    working_days = 0
    while current_date <= end_date:
        if current_date.weekday() < 5:  # Monday to Friday
            working_days += 1
        current_date += timedelta(days=1)
    return working_days

# Function to generate Gantt Chart using Matplotlib and Seaborn
def generate_gantt_chart_matplotlib(df, title, subtitle):
    fig, ax = plt.subplots(figsize=(12, 6))

    # Plot Gantt chart
    sns.set_theme(style="whitegrid")
    ax = sns.barplot(x="Start", y="Task", hue="Domain", data=df, palette="pastel", edgecolor="black")

    # Format x-axis to show the first day of each month
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

    # Set the title and subtitle
    plt.title(title, fontsize=16, fontweight='bold')
    plt.suptitle(subtitle, y=1.05, fontsize=12, ha='center')

    plt.show()

# Function to generate Gantt Chart using Plotly Express
def generate_gantt_chart_plotly(df, title, subtitle):
    fig = px.timeline(df, x_start="Start", x_end="End", y="Task", color="Domain", labels={"Task": "Activity"},
                      category_orders={"Domain": ["Takaful", "Conventional"]}, title=title, text="Task")

    # Set the title and subtitle
    fig.update_layout(
        title_text=title,
        title_font_size=16,
        title_font=dict(weight='bold'),
        title_y=0.95,
        title_x=0.5,
        xaxis_title_text=subtitle,
        xaxis_title_font_size=12,
        xaxis_title_font=dict(weight='normal')
    )

    fig.show()

# Define the project stages with their start and end dates
project_stages_takaful = {
    'Contract signing - Start of the Project': {'start': '2023-11-21', 'end': '2023-11-21'},
    'Planning Stage': {'start': '2023-11-22', 'end': '2023-11-28'},
    'Analysis Stage': {'start': '2023-11-29', 'end': '2024-02-14'},
    'Development Stage (POS)': {'start': '2023-12-01', 'end': '2024-02-15'},
    'Testing Phase (POS)': {'start': '2024-02-16', 'end': '2024-03-31'},
    'Development Stage (ILAS)': {'start': '2023-12-01', 'end': '2024-03-31'},
    'Testing Phase (ILAS)': {'start': '2024-04-02', 'end': '2024-05-01'},
    'Deployment Phase': {'start': '2024-05-15', 'end': '2024-05-30'},
    'Go Live Phase': {'start': '2024-05-31', 'end': '2024-05-31'},
    'Post-Launch Support': {'start': '2024-06-01', 'end': '2025-02-28'}
}

# Define the project stages with their start and end dates for Conventional (Phase 2)
project_stages_conventional = {
    'Deployment Phase': {'start': '2024-06-01', 'end': '2024-11-30'},
}

# Create DataFrames for Takaful and Conventional
df_takaful = pd.DataFrame(project_stages_takaful.items(), columns=['Task', 'Dates'])
df_conventional = pd.DataFrame(project_stages_conventional.items(), columns=['Task', 'Dates'])

# Split Dates into Start and End columns
df_takaful[['Start', 'End']] = df_takaful['Dates'].apply(lambda x: pd.Series([x['start'], x['end']]))
df_conventional[['Start', 'End']] = df_conventional['Dates'].apply(lambda x: pd.Series([x['start'], x['end']]))

# Drop unnecessary columns
df_takaful = df_takaful.drop(['Dates'], axis=1)
df_conventional = df_conventional.drop(['Dates'], axis=1)

# Add Domain column
df_takaful['Domain'] = 'Takaful'
df_conventional['Domain'] = 'Conventional'

# Calculate working days for each task
df_takaful['Working Days'] = df_takaful.apply(lambda row: calculate_working_days(datetime.strptime(row['Start'], "%Y-%m-%d"),
                                                                              datetime.strptime(row['End'], "%Y-%m-%d")), axis=1)
df_conventional['Working Days'] = df_conventional.apply(lambda row: calculate_working_days(datetime.strptime(row['Start'], "%Y-%m-%d"),
                                                                                        datetime.strptime(row['End'], "%Y-%m-%d")), axis=1)

# Generate Gantt charts
title = "Gantt Chart for Development Timeline for POS and Back-end"
subtitle = "Takaful and Conventional Domain"

# Matplotlib and Seaborn Gantt Chart for Takaful
generate_gantt_chart_matplotlib(df_takaful, title, subtitle)

# Matplotlib and Seaborn Gantt Chart for Conventional (Phase 2)
generate_gantt_chart_matplotlib(df_conventional, title, subtitle)

# Plotly Express Gantt Chart for Takaful
generate_gantt_chart_plotly(df_takaful, title, subtitle)

# Plotly Express Gantt Chart for Conventional (Phase 2)
generate_gantt_chart_plotly(df_conventional, title, subtitle)
```
